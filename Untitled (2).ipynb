```python

def plot_count(df, x_var, axis = 'both', palette = 'Blues', ax = None):
    sns.countplot(data = df,\
                  x = x_var,\
                  palette = palette,\
                  edgecolor = 'black',\
                  linewidth = 1.5,\
                  width = 0.7,\
                  saturation = 0.8,\
                  ax = ax
                )
    ax.set_title(f'~Distribution of {x_var}~',\
              fontsize = 16,\
              fontweight = 'bold')
    ax.set_xlabel(x_var,\
               fontsize = 14)
    ax.set_ylabel('Values',\
               fontsize = 14)
    ax.grid(visible = True,\
             alpha = 0.6,\
             linestyle = '--',\
             color = 'black',\
             axis = axis)
```

```python

def plot_line(x_var, y_var, hue, xt = None, yt = None, group = None, palette = None, axis = 'both', ax = None):
    '''
    the function will create a customize line plot according to conditions.
    '''
    sns.lineplot(data = group,\
                 x = x_var,\
                 y = y_var,\
                 hue = hue,\
                 palette = palette,\
                 markers = True,\
                 dashes = False,\
                 linewidth = 2,\
                 legend = 'full',\
                 err_style = "band",\
                 markersize = 6,\
                 units = None,\
                 ax = ax
                )
    ax.set_title(f'~Charges vs {x_var} & {hue}',\
              fontsize = 16,\
              fontweight = 'bold')
    ax.set_xlabel(x_var,\
               fontsize = 14)
    ax.set_ylabel('Charges',\
               fontsize = 14)
    ax.set_xticks(xt)
    ax.set_yticks(yt)
    ax.grid(visible = True,\
             alpha = 0.6,\
             linestyle = '--',\
             color = 'black',\
             axis = axis)
```

```python
def calculate_regression_metrics(y_true: pd.Series, y_pred: pd.Series) -> Tuple[float, float, float, float, float]:  
    mse = mean_squared_error(y_true, y_pred)  
    mae = mean_absolute_error(y_true, y_pred)  
    rmse = np.sqrt(mse)  
    r2 = r2_score(y_true, y_pred)  

    with np.errstate(divide='ignore', invalid='ignore'):  
        mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100  
        mape = np.nan_to_num(mape) 

    return mse, mae, rmse, r2, mape  


data = {'actual': [10, 20, 30, 40, 50], 'predicted': [12, 18, 32, 38, 55]}  
df = pd.DataFrame(data)  

mse, mae, rmse, r2, mape = calculate_regression_metrics(df['actual'], df['predicted'])  

print(f"Mean Squared Error (MSE): {mse:.2f}")  
print(f"Mean Absolute Error (MAE): {mae:.2f}")  
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")  
print(f"R-squared (R2): {r2:.2f}")  
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```
